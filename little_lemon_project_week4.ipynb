```python
# Task 1: Establishing a connection

from mysql.connector.pooling import MySQLConnectionPool
from mysql.connector import Error

# Database configurations
dbconfig = {"database" : "little_lemon_db", "user" : "root", "password" : ""}

try:
    pool = MySQLConnectionPool(pool_name = "pool_a",
    pool_size = 2,
    **dbconfig)
    print("The connection pool is created with a name: ", pool.pool_name)
    print("The pool size is:", pool.pool_size)
except Error as er:
    print("Error code:", er.errno)
    print("Error message:", er.msg)
```

```python
# Obtain a connection from pool_a and create a cursor object to communicate with the database
print("Getting a connection from the pool.")
connection_one = pool.get_connection()
cursor = connection_one.cursor()
print("Connection one created.")
```

```python
# Task 2: Implement a stored procedure called PeakHours

peakhours_procedure = """
CREATE PROCEDURE PeakHours()
BEGIN
SELECT HOUR(BookingSlot) AS Hour,
COUNT(BookingSlot) AS No_of_Bookings
FROM Bookings
GROUP BY Hour
ORDER BY No_of_Bookings DESC;
END"""

# Execute the stored procedure
cursor.execute(peakhours_procedure)

# Invoke callproc to call stored procedure
cursor.callproc("PeakHours")

# Fetch the results
results = next(cursor.stored_results())
dataset = results.fetchall()

# Extract the names of the columns
for column_id in cursor.stored_results():
    columns = [ column[0] for column in column_id.description ]

# Print the sorted data using a for loop
for data in dataset:
    print(data)
```

```python
# Task 3: Implement a stored procedure called GuestStatus

gueststatus_proc = """CREATE PROCEDURE GuestStatus()
BEGIN

SELECT
Bookings.BookingID AS OrderNumber,
CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) AS GuestName,

Role AS Employee,

CASE
WHEN Role = "Manager" OR Role = "Assistant Manager" THEN "Ready to pay"
WHEN Role = "Head Chef" THEN "Ready to serve"
WHEN Role = "Assistant Chef" THEN "Preparing order"
WHEN Role = "Head Waiter" THEN "Order served"
ELSE "In queue"
END AS Status

FROM Bookings
LEFT JOIN Employees
ON Bookings.EmployeeID = Employees.EmployeeID;

END"""

# Execute the stored procedure
cursor.execute(gueststatus_proc)

# Invoke callproc to call stored procedure
cursor.callproc("GuestStatus")

# Fetch the results
results = next(cursor.stored_results())
dataset = results.fetchall()

# Extract the names of the columns
for column_id in cursor.stored_results():
    columns = [ column[0] for column in column_id.description ]

# Print the sorted data using a for loop
for data in dataset:
    print(data)
```

```python
# Close the connection to return it back to the pool
print("Returning the connection back to the pool.")
connection_one.close()
print("The connection is placed back into the pool for the next user to connect.")
```
